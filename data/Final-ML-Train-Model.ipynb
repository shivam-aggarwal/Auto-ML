```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df
```

```python
dummies = pd.get_dummies(df.Algorithm, prefix = 'Algorithm_')
for col in list(dummies.columns):
    df[col] = dummies[col]
del df['Algorithm']
```

```python
df
```

```python
df['score'] = df['testing_score'].copy()
del df['testing_score']
del df['best_score']
del df['training_score']
```

```python
listDataset = list(set(df['dataset']))
```

```python
listDataset
```

```python
training_set = set(random.choices(listDataset, k = 7))
```

```python
testing_set = set(listDataset) - training_set
```

```python
training_set, testing_set
```

```python
training_set = {'SomervilleHappinessSurvey2015',
  'abalone',
  'column_2C',
  'ecoli',
  'mammographic_masses',
  'user'}
testing_set = {'column_3C', 'data_banknote_authentication', 'seed', 'transfusion'}
```

```python
train_df = df[df.dataset.isin(training_set)].copy()
```

```python
test_df = df[df.dataset.isin(testing_set)].copy()
```

```python
train_df_dataset = list(train_df.dataset)
del train_df['dataset']
```

```python
test_df_dataset = list(test_df.dataset)
del test_df['dataset']
```

```python
X_train = train_df.iloc[:, :-1].values
X_test = test_df.iloc[:, :-1].values
```

```python
Y_train = train_df.iloc[:, -1].values
Y_test = test_df.iloc[:, -1].values
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(max_depth=5, min_samples_leaf=2, random_state=42)
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
Y_test_pred = model.predict(X_test)
```

```python
Y_train_pred = model.predict(X_train)
```

```python
list(Y_train_pred)
```

```python
train_df['dataset'] = train_df_dataset
train_df['predicted_score'] = list(Y_train_pred)
```

```python
test_df['dataset'] = test_df_dataset
test_df['predicted_score'] = list(Y_test_pred)
```

```python
test_df[['dataset', 'score', 'predicted_score']]
```
