```python
raw_data = loadarff('column_3C_weka.arff')
df_data = pd.DataFrame(raw_data[0])
```

```python
classes = list(set(df_data['class']))
```

```python
classes
```

```python
classes = [b'Normal', b'Hernia', b'Spondylolisthesis']
```

```python
target = []
for i in range(len(df_data)):
    val = df_data.at[i, 'class']
    res = 0
    for j in range(len(classes)):
        if classes[j] == val:
            res += j
    assert(j <= 2)
    target.append(res)
```

```python
del df_data['class']
```

```python
df_data['target'] = target
```

```python
df_data.head()
```

```python
df_data.to_csv('column_3C.csv', index = False)
```
