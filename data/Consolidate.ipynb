```python
from sklearn.decomposition import PCA
import pandas as pd
import os
import json
import numpy as np
```

```python
META_DIR = "results-metadata/"
DIR = "results/final/"
```

```python
os.listdir(META_DIR)
```

```python
os.listdir(DIR)
```

```python
import json
```

```python
data = []
for file_name in os.listdir(META_DIR):
    with open(META_DIR + file_name, "r") as f:
        row = json.load(f)
        row['dataset'] = file_name.split('-')[0]
        data.append(row)
```

```python
metaDf = pd.DataFrame(data=data)
```

```python
data = []
for file_name in os.listdir(DIR):
    with open(DIR + file_name, "r") as f:
        row = json.load(f)
        row['dataset'] = file_name.split('-')[0]
        data.append(row)
```

```python
detailsDf = pd.DataFrame(data=data)
```
